```python
def extract_lines(input_file_path, start_line, end_line, output_file_path):
    try:
        with open(input_file_path, 'r') as input_file, open(output_file_path, 'w') as output_file:
            lines = input_file.readlines()
            # Adjusting start_line and end_line to be within the valid range
            start_line = max(1, min(start_line, len(lines)))
            end_line = max(start_line, min(end_line, len(lines)))

            # Writing the selected lines to the output file
            output_file.writelines(lines[start_line - 1:end_line])

        print(f"Lines {start_line} to {end_line} extracted and saved to {output_file_path}")

    except Exception as e:
        print(f"Error: {e}")


import pandas as pd
def write_excel(items, path):
    if len(items) == 0:
        return

    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(items)

    # Write DataFrame to Excel
    with pd.ExcelWriter(path, engine='xlsxwriter') as writer:
        df.to_excel(writer, index=False, sheet_name='Sheet1')
```

```python
import requests
from bs4 import BeautifulSoup
import time
def x(url):
  headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36"}
  r = requests.get(url, headers=headers)
  time.sleep(2)
  return r

def x2(r):
  if not r.ok:
    print('Status Code:', r.status_code)
    raise Exception('Failed to fetch')

  soup = BeautifulSoup(r.text, 'html.parser')
  time.sleep(1)
  return soup
```

```python
import re

def x3(soup, url, title, price, shipping_cost, number_sold):
    brand=""
    manufacturer=""
    manufacturer_num=""
    oem=""
    type1=""
    num_sold = number_sold
    if num_sold == "None":
      quantity_sold = soup.find('div', class_="d-quantity__availability")
      if quantity_sold:
        both = quantity_sold.text.strip()
        split = both.split(' / ')
        if len(split)>1:
          num_sold = split[1]
        else:
          if "verkauft" in both:
            num_sold = both
    if num_sold == "None":
      return None
    layout = soup.find('div', class_="ux-layout-section-evo ux-layout-section--features")
    rows = layout.find_all('div', class_="ux-layout-section-evo__col")

    for row in rows:
      header_label=""
      header = row.find('dt', class_="ux-labels-values__labels")
      value = row.find('dd', class_="ux-labels-values__values")
      if header:
        header_label = header.text.strip()
      if value:
        value =value.text.strip()
      if header_label == "Marke":
        brand = value
      elif header_label == "Hersteller":
        manufacturer = value
      elif header_label == "Herstellernummer":
        manufacturer_num = value
      elif "OEM" in header_label:
        oem = value
      elif header_label == "Produktart":
        type1 = value

    #Return a Dictionary
    return{
        'Title': title,
        'Price': price,
        'Shipping Cost': shipping_cost,
        'Sold number': num_sold,
        'Brand': brand,
        'Manufacturer': manufacturer,
        'Manufacturer Number': manufacturer_num,
        'OEM': oem,
        'type': type1,
        'url': url,
    }
```

```python
from tqdm import tqdm
def x4(file_path):
    # Read URLs from the file
    with open(file_path, 'r') as file:
        lines = file.read().splitlines()

    # Scrape data for each URL
    results = []
    #counter=0
    for line in tqdm(lines):
        # Split the line by " | " to separate the name and URL
        url, title, price, shipping_cost, number_sold  = line.split(" \ ")
        try:
            r=x(url)
            soup=x2(r)
            data=x3(soup, url, title, price, shipping_cost, number_sold)
            if data:
              results.append(data)
              print(data)
            else:
              print("None Sold")
            time.sleep(2)

        except Exception as e:
            print(f"Error scraping data for URL: {url}\nError: {e}")
            # Write the error URL to errors.txt
            with open('errors_ebay.txt', 'a') as error_file:
                error_file.write(line + '\n')

    return results
```

```python
extract_lines('ebay_data.txt', 2501, 3000, 'file.txt')
# Example usage:
urls_file_path = 'file.txt'
output_csv_path = 'ebay6.xlsx'

# Get scraped data for each URL
scraped_data = x4(urls_file_path)

# Write the results to a CSV file
write_excel(scraped_data, output_csv_path)
```
